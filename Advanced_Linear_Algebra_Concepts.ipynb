Determinants and Inverse of a Matrix
* Determinants
  * Scalar value that provides information about a matrix‚Äôs properties
  * Only for square matrices
  * det(A) = 0, the matrix A is singular
  * det(A) != 0, A is invertible
  * Geometric Interpretation
    * For a 2√ó2 matrix, the determinant represents the scaling factor of the area
formed by its column vectors
  * Formula for 2 √ó 2 Matrix:
    image.png

```python
import numpy as np

A = np.array([[2,3],[4,5]])
determinant = np.linalg.det(A)
print("Determinant of A:\t",determinant)
```

Eigenvalues and Eigenvectors
* What are Eigenvalues and Eigenvectors?
* If ùê¥ ‚ãÖ ùë£ = ùúÜ ‚ãÖ ùë£, then
  * ùë£ is an eigenvector
  * Œª is the eigenvalue
* Geometric Interpretation
  * Eigenvectors point in the direction where the matrix transformation stretches or compresses vectors
  * Eigenvalues indicate the factor of stretching or compression
* Properties
  * Matrix of size ùëõ √ó ùëõ has ùëõ n eigenvalues and eigenvectors
  * Eigenvalues can be real or complex
  * For a symmetric matrix, eigenvalues are always real.
* Computing Eigenvalues and Eigenvectors in NumPy

```python
B = np.array([[1,2], [3,4]])
eigval, eigvec = np.linalg.eig(B)
print("EigVal: \n",eigval)
print("EigVec: \n",eigvec)
```

Introduction to Matrix Decomposition
* What is Matrix Decomposition?
  * Process of breaking a matrix into simpler components to analyze or solve
problems
* Singular Value Decomposition (SVD)
  * SVD decomposes a matrix ùê¥ A into three matrices: ùê¥ = ùëà ‚ãÖ Œ£ ‚ãÖ ùëâT
    * U: Left singular vectors (orthogonal matrix)
    * Œ£: Diagonal matrix of singular values (non-negative).
    * ùëâT: Right singular vectors (orthogonal matrix).
* Applications of SVD
* Computing SVD in NumPy

```python
C = np.array([[1,2],[3,4]])

U, S, Vt = np.linalg.svd(C)
print("U: \n",U)
print("Singular Values: \n",S)
print("V Transpose: \n",Vt)
```

Hands-On Exercises

```python
Matrix = np.array([[1,4],[9,6]])
U, S, Vt = np.linalg.svd(Matrix)
print("U: \n",U)
print("Singular Values: \n",S)
print("V Transpose: \n",Vt)

#Reconstruct
Sigma = np.zeros((2,2))
np.fill_diagonal(Sigma,S)
reconstrucuted = U @ Sigma @ Vt
print("Reconstructed Matrix: \n", reconstrucuted)
```
