```python
#Exercise 1: Clean a Dataset by Handling Missing Values and Renaming Columns
import numpy as np
import pandas as pd

#Created sample data set
data = {
    "Name": ["Alice","Bob","John",np.nan],
    "Age":[20,22,np.nan,19],
    "Score": [88,np.nan,78,np.nan]
}
df = pd.DataFrame(data)
print("Original Data:\n",df,"\n")

#Handling missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Score"] = df["Score"].fillna(df["Score"].interpolate())
print("Clean Dataset:\n",df,"\n")

# Renaming columns
df = df.rename(columns= {"Name":"Student_Name","Score":"Marks"})
print("After Renaming:\n",df,"\n")
```

```python
#Exercise 2: Merge Two Datasets and Perform Data Transformations

import numpy as np
import pandas as pd

#creating 2 sample datasets
df1 = pd.DataFrame({
    "ID" : [1,2,3],
    "Name" : ["Alice","Bob","John"],
    "Age" : [20,22,21]
})

df2 = pd.DataFrame({
    "ID" : [1,2,3],
    "Score" : [45,60,57]
})

print("\n Data Set 1: \n",df1)
print("\n Data Set 2: \n",df2)

# Merging 2 datasets
merged = pd.merge(df1,df2,how="inner",on="ID")
print("\nMerged Data set:\n",merged)

# Data transformation
merged["Score_%"] = (merged["Score"] / 200) * 100
print("\nTransformed Data set:\n",merged)
```

* Merge three datasets and analyze relationships between them
* Convert categorical data to numerical using one-hot encoding

```python
# Drop columns with more than 50% missing values

import numpy as np
import pandas as pd

# Creating a sample dataset
df = pd.DataFrame({
    "ID" : [1,2,3,4,5],
    "Age" : [20,np.nan,22,np.nan,np.nan],
    "Gender" : ["M","F",np.nan,"M",np.nan],
    "Marks" : [85,np.nan,78,70,np.nan],
    "Attendance" : [90,85,np.nan,np.nan,np.nan],
    "City" : ["Delhi",np.nan,np.nan,np.nan,np.nan]
})
print("\nOriginal dataset\n",df)

# Main logic

total_rows = len(df)
missing_count = df.isnull().sum()
half_data = total_rows/2
columns_to_drop = missing_count[missing_count > half_data].index
clean_data = df.drop(columns=columns_to_drop)
print("\n Cleaned Data: \n",clean_data)
```

```python
# Merge three datasets and analyze relationships between them

import numpy as np
import pandas as pd

# dataset 1
df1 = pd.DataFrame({
    "ID" : [1,2,3,4],
    "Student_ID" : ["Alice","Bob","John","Tom"]
})

# dataset 2
df2 = pd.DataFrame({
    "ID" : [1,2,3,4],
    "Marks" : [80,50,78,59]
})

# dataset 3
df3 = pd.DataFrame({
    "ID" : [1,2,3,4],
    "Attendance" : [79,86,74,70]
})

first_merge = pd.merge(df1,df2,on="ID",how="left")
second_merge = pd.merge(first_merge,df3,on="ID",how="left")
print(second_merge)
```
